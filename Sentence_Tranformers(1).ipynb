```python
model = SentenceTransformer('stsb-roberta-large')
```

```python
generator = pipeline('text-generation', model='gpt2-large')
```

```python
ltop=["Linear Regression","Supervised Classification","Unsupervised Clustering","Dimensionality Reduction","Perceptron Regression","Support Vector Machines","Reinforcement Learning"]  
tpoc=[]
```

```python
language = "en"
max_ngram_size =5
deduplication_threshold = 0.9
numOfKeywords = 3
tpr=0
custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
for sn in ltop:
    dn=generator(sn, max_length=12, num_return_sequences=500)
    finalkeywords=[]
    for i in dn:
        text=i['generated_text']
        if(tpr<=2):
            print("GPT2 output: "+text)
        keywords = custom_kw_extractor.extract_keywords(text)
        for i in keywords:
            finalkeywords.append(i[0])
            if(tpr<=2):
                print("YAKE Outputs : "+i[0])
        tpr=tpr+1
    fs=[' '.join(finalkeywords)]
    tpoc.append(fs)
```

```python
resem=[]
for i in res:
    embedding1 = model.encode(i, convert_to_tensor=True)
    resem.append(embedding1)
topem=[]
for j in tpoc:
    embedding2 = model.encode(j, convert_to_tensor=True)
    topem.append(embedding2)
polyl=[]
for i in resem:
    r1=[]
    for j in topem:
        temp=util.pytorch_cos_sim(i, j)
        r1.append(temp)
    polyl.append(r1)
polyl1=[]
for i in topem:
    r2=[]
    for j in topem:
        temp=util.pytorch_cos_sim(i, j)
        r2.append(temp)
    polyl1.append(r2)
```
